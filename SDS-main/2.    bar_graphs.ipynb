```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
df= pd.read_excel('clean_ver.xlsx', index_col=0) 
df['year'] = [d.year for d in df.Date]
df['month'] = [d.strftime('%b') for d in df.Date]
df1=df[(df.month != 'Oct')]
df1=df1[(df1.month != 'Nov')]
df1=df1[(df1.month != 'Dec')]
df1=df1[(df1.month != 'Sep')]
df1=df1[(df1.month != 'Aug')]
```

```python
sns.countplot(x="year", hue="AQI_Bucket", data=df1)
```

```python
x2= df[['PM2.5','City']].groupby(["City"]).median().sort_values(by='PM2.5',ascending=False).reset_index()
```

```python
g=sns.barplot(x=x2['City'], y= x2['PM2.5'])
g.set_xticklabels(g.get_xticklabels(), rotation=45)
```
